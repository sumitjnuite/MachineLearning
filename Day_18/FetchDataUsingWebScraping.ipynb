```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.amazon.in/s?k=macbook+m1&page=2"
headers = {
    'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36",
    'Referer': 'https://www.google.com',  # Fake a referrer
}

webpage = requests.get(url,headers=headers).text
soup = BeautifulSoup(webpage,'lxml')
print(soup.prettify())
```

```python

# Scraping title and price data from Amazon website for a search keyword macbook m1 

productsDiv=[]
title = []
price = []
for j in range(11):
    headers = {
        'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36",
        'Referer': 'https://www.google.com',  # Fake a referrer
    }
    webpage = requests.get("https://www.amazon.in/s?k=macbook+m1&page={}".format(j),headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    # print(soup.prettify())
    
    productsDiv.append(soup.find_all('div',class_="a-section a-spacing-small a-spacing-top-small"))


```

```python
for products in productsDiv:
    for i in products:
        title.append(i.find('h2',class_="a-size-medium a-spacing-none a-color-base a-text-normal").text)
        price.append(i.find('span',class_="a-price-whole").text)
        # print(i.text.strip())
```

```python
df2=pd.DataFrame({'title':title,
   'price':price
   })
```
